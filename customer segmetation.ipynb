```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('E:\excel datasets\Mall_Customers.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
x = df.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss = []
```

```python
for i in range (1,11):
    kmeans = KMeans(n_clusters = i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('number of clusters')
plt.ylabel('WCSS values')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 5,init='k-means++',random_state = 0)
```

```python
y_kmeans = kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0,0], x[y_kmeans == 0,1],s = 80,c= "red",label='customer 1')
plt.scatter(x[y_kmeans == 1,0], x[y_kmeans == 1,1],s = 80,c= "yellow",label='customer 2')
plt.scatter(x[y_kmeans == 2,0], x[y_kmeans == 2,1],s = 80,c= "black",label='customer 3')
plt.scatter(x[y_kmeans == 3,0], x[y_kmeans == 3,1],s = 80,c= "orange",label='customer 4')
plt.scatter(x[y_kmeans == 4,0], x[y_kmeans == 4,1],s = 80,c= "blue",label='customer 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="magenta",label='centroids')
plt.title('customer segmentation')
plt.xlabel('annual income ($k)')
plt.ylabel('spending score (1-100)')
plt.legend()
plt.show()
```
